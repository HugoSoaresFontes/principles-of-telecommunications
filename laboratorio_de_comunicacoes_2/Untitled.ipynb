```python
import math
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error 
from cmath import phase
```

```python
R = 2700
C = 1e-9

from numpy import genfromtxt
data = genfromtxt('../../controle2/somador.csv', delimiter=',')
v = data[:, 1]
t = data[:, 0]
fft = np.fft.fft(v)

T = (t[1] - t[0])
f = np.linspace(0.0, 1.0/T, len(t))
resposta_filtro = 1.0 / np.sqrt((2.0 * np.pi * 2.7e3 * 1e-9 * f)**2 + 1)

sinal_filtrado_frequencia = fft * resposta_filtro

%matplotlib notebook
plt.figure(figsize=(10,6))
plt.plot(f[:200], fft[:200]/max(fft), 'b', linewidth=1, label="Final sem filtro")
plt.plot(f[:200], sinal_filtrado_frequencia[:200]/(fft), 'r', linewidth=1, label="Sinal Filtrado")
plt.legend()
plt.show()
```

```python
sinal_filtrado_tempo = np.fft.ifft(sinal_filtrado_frequencia)
%matplotlib notebook
plt.figure(figsize=(10,6))
#plt.plot(f[:200], fft[:200]/max(fft), 'b', linewidth=1, label="Final sem filtro")
plt.plot(t, v, 'b', linewidth=1, label="Sinal sem filtro")
plt.plot(t, sinal_filtrado_tempo, 'r', linewidth=1, label="Sinal Filtrado")
plt.legend()
plt.show()

```
